```python
df = pd.read_csv('amazon kindle.csv' , sep = ',') # Reload the original dataframe to ensure all columns are present
df = df[['Name', 'Review', 'Price($)']]
df.head(10)
```

```python
df.sort_values('Review', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))

Amozonkindle_counts = df['Review'].value_counts()

sns.barplot(
    x=Amozonkindle_counts.values,
    y=Amozonkindle_counts.index
)

plt.title('Gêneros Mais Frequentes')
plt.xlabel('Quantidade')
plt.ylabel('Gênero')

plt.show()
```

```python
top_10_produtos = df.sort_values('Review', ascending=False).head(10)

plt.figure(figsize=(12, 7))

# Criar o gráfico de barras
plt.bar(x=top_10_produtos['Name'], height=top_10_produtos['Review'], color='skyblue')

plt.xlabel('Nome do Produto')
plt.ylabel('Avaliação')
plt.title('Top 10 Produtos por Avaliação')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
